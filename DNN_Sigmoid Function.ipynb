```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Negative input values
negative_inputs = [-100, -10, -5, -1]
negative_outputs = [sigmoid(x) for x in negative_inputs]

print("Sigmoid Function (Negative Inputs):")
for x, y in zip(negative_inputs, negative_outputs):
    print(f"Sigmoid({x}) = {y}")

# Zero and positive input values
positive_inputs = [0, 1, 5, 10, 100]
positive_outputs = [sigmoid(x) for x in positive_inputs]

print("\nSigmoid Function (Zero and Positive Inputs):")
for x, y in zip(positive_inputs, positive_outputs):
    print(f"Sigmoid({x}) = {y}")

# Plotting
x_values = np.linspace(-10, 10, 100)
y_values = sigmoid(x_values)
plt.plot(x_values, y_values, label="Sigmoid")
plt.scatter(negative_inputs, negative_outputs, color='red', label="Negative Inputs")
plt.scatter(positive_inputs, positive_outputs, color='blue', label="Positive Inputs")
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the Tanh function
def tanh(x):
    return np.tanh(x)

# Negative input values
negative_inputs = [-10, -5, -1]
negative_outputs = [tanh(x) for x in negative_inputs]

print("Tanh Function (Negative Inputs):")
for x, y in zip(negative_inputs, negative_outputs):
    print(f"Tanh({x}) = {y}")

# Zero and positive input values
positive_inputs = [0, 1, 5, 10]
positive_outputs = [tanh(x) for x in positive_inputs]

print("\nTanh Function (Zero and Positive Inputs):")
for x, y in zip(positive_inputs, positive_outputs):
    print(f"Tanh({x}) = {y}")

# Plotting
x_values = np.linspace(-10, 10, 100)
y_values = tanh(x_values)
plt.plot(x_values, y_values, label="Tanh")
plt.scatter(negative_inputs, negative_outputs, color='red', label="Negative Inputs")
plt.scatter(positive_inputs, positive_outputs, color='blue', label="Positive Inputs")
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the ReLU function
def relu(x):
    return max(0, x)

# Negative input values
negative_inputs = [-10, -5, -1]
negative_outputs = [relu(x) for x in negative_inputs]

print("ReLU Function (Negative Inputs):")
for x, y in zip(negative_inputs, negative_outputs):
    print(f"ReLU({x}) = {y}")

# Zero and positive input values
positive_inputs = [0, 1, 5, 10]
positive_outputs = [relu(x) for x in positive_inputs]

print("\nReLU Function (Zero and Positive Inputs):")
for x, y in zip(positive_inputs, positive_outputs):
    print(f"ReLU({x}) = {y}")

# Plotting
x_values = np.linspace(-10, 10, 100)
y_values = [relu(x) for x in x_values]
plt.plot(x_values, y_values, label="ReLU")
plt.scatter(negative_inputs, negative_outputs, color='red', label="Negative Inputs")
plt.scatter(positive_inputs, positive_outputs, color='blue', label="Positive Inputs")
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```
